```python
#Crate Dataframe to save train_dataset and test_dataset

train_dataset = pd.DataFrame()
train_dataset['title + author'] = train['title'] + ' ' + train['author']
train_dataset['label'] = train['label']

test_dataset = pd.DataFrame()
test_dataset['title + author'] = test['title'] + ' ' + test['author']
test_dataset['label'] = test['label']
```

```python
def DropNaN(df):
  #dropna 
  df = df.dropna()
  #reset index
  df.reset_index(inplace=True)
  return df
```

```python
#Drop Nan

train_dataset = DropNaN(train_dataset)
train_dataset.info()
```

```python
test_dataset = DropNaN(test_dataset)
test_dataset.info()
```

```python
train_dataset['title + author'] = train_dataset['title + author'].astype(str)
test_dataset['title + author'] = test_dataset['title + author'].astype(str)
```

```python
#preprocessing def

def Preprocessing(text):
  corpus = []
  for i in range(0, len(text)):
    data_point = re.sub('[^a-zA-Z0-9]', ' ', text[i])
    data_point = data_point.lower()
    data_point = data_point.split()
    data_point = [word for word in data_point if not word in stopwords.words('english')]
    data_point = ' '.join(data_point)
    corpus.append(data_point)
  return corpus
```

```python
train_dataset['title + author'] = Preprocessing(train_dataset['title + author'])
test_dataset['title + author'] = Preprocessing(test_dataset['title + author'])

```
