```python

n =int(input("Nhap so phan tu cua day: "))
list = []
for x in range(n):
    a = int(input())
    list.append(a)

def tinh_trung_binh(list):
    # s = sum(int(x) for x in list)
    s = sum(list)
    n=len(list)
    avg= float(s/n)
    return avg
print(tinh_trung_binh(list))
```
