```python
import math 
import matplotlib.pyplot as plt


def data_exp(x):
  return  math.exp(x)

x_data = [x/10 for x in range(1,100)]
# x_data = [x for x in range(1,10)]
y_data = [data_exp(x) for x in x_data]

plt.figure(figsize = (6,4)) #kích cỡ khung của đô thị
plt.plot(x_data, y_data, label = 'e^x') # hiển thị thuộc tính của đồ thị
plt.legend() #show hiển thị data
```
